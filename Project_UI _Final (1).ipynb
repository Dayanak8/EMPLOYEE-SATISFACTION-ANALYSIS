```python
from tkinter import *
from sklearn.ensemble import AdaBoostRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.multioutput import MultiOutputRegressor
import tkinter as tk
from tkinter import ttk
import numpy as np
from sklearn.metrics import mean_squared_error

# Sample data
new_data = {'Face_stress': np.random.rand(100), 'Satisfaction_of_salary': np.random.rand(100)}
X = np.array(new_data['Face_stress']).reshape(-1, 1)
y = np.array(new_data['Satisfaction_of_salary'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)

# MSE variables
mse_dict = {'SGDRegressor': None, 'AdaBoostRegressor': None,
            'XGBRegressor': None, 'MultiOutputRegressor': None}

# Regressor objects
Reg1 = XGBRegressor()  # Default regressor
Reg2 = AdaBoostRegressor()  # Default regressor
Reg3 = SGDRegressor()  # Default regressor
Reg4 = MultiOutputRegressor(SGDRegressor())  # Default regressor

def calculate_mse(algorithm, model, X_test_data, y_test_data):
    predictions = model.predict(X_test_data.reshape(-1, 1))
    mse = mean_squared_error(y_test_data, predictions)
    mse_dict[algorithm] = mse
    print(f"{algorithm} MSE: {mse}")

def ml_tkinter_gui():
    global root
    root = Tk()

    # Creating canvas to show the result
    canvas1 = tk.Canvas(root, width=500, height=450)
    canvas1.pack()

    # Function to close the window
    def close_window():
        root.destroy()

    # Add button
    button = tk.Button(text="Reference & Restart", command=refresh, bg='green')
    button.place(x=50, y=350)

    # Add button
    button = tk.Button(text="Click and Quit", command=close_window, bg='red')
    button.place(x=310, y=350)

    # Algorithm selection
    Algorithm_list = ['SGDRegressor', 'AdaBoostRegressor', 'XGBRegressor', 'MultiOutputRegressor']
    Algorithm = StringVar()
    Algorithm.set(Algorithm_list[2])
    Algorithm_combobox = ttk.Combobox(root, width=20, height=10, textvariable=Algorithm)
    Algorithm_combobox['values'] = Algorithm_list
    Algorithm_combobox.place(x=150, y=50)

    # MSE Matrices
    def score():
        global mse1, Reg3, Reg2, Reg1, Reg4
        if Algorithm.get() == 'SGDRegressor':
            Reg3 = SGDRegressor()
            Reg3.fit(X_train.reshape(-1, 1), y_train)
            mse1 = mean_squared_error(y_test, Reg3.predict(X_test.reshape(-1, 1)))
            calculate_mse('SGDRegressor', Reg3, X_test, y_test)
        elif Algorithm.get() == 'AdaBoostRegressor':
            Reg2 = AdaBoostRegressor()
            Reg2.fit(X_train.reshape(-1, 1), y_train)
            mse1 = mean_squared_error(y_test, Reg2.predict(X_test.reshape(-1, 1)))
            calculate_mse('AdaBoostRegressor', Reg2, X_test, y_test)
        elif Algorithm.get() == 'XGBRegressor':
            Reg1 = XGBRegressor()
            Reg1.fit(X_train.reshape(-1, 1), y_train)
            mse1 = mean_squared_error(y_test, Reg1.predict(X_test.reshape(-1, 1)))
            calculate_mse('XGBRegressor', Reg1, X_test, y_test)
        elif Algorithm.get() == 'MultiOutputRegressor':
            Reg4 = MultiOutputRegressor(SGDRegressor())
            Reg4.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))
            mse1 = mean_squared_error(y_test, Reg4.predict(X_test.reshape(-1, 1)))
            calculate_mse('MultiOutputRegressor', Reg4, X_test, y_test)

        Prediction_result = ('MSE: ', mse1)
        label_Prediction = tk.Label(root, text=Prediction_result, bg='orange')
        canvas1.create_window(100, 250, window=label_Prediction)

    # Test the data model
    def test_score():
        if 'mse1' in globals() and mse1 is not None:
            if Algorithm.get() == 'MultiOutputRegressor':
                y_pred = Reg4.predict(X_test.reshape(-1, 1))
            elif Algorithm.get() == 'SGDRegressor':
                y_pred = Reg3.predict(X_test.reshape(-1, 1))
            elif Algorithm.get() == 'AdaBoostRegressor':
                y_pred = Reg2.predict(X_test.reshape(-1, 1))
            elif Algorithm.get() == 'XGBRegressor':
                y_pred = Reg1.predict(X_test.reshape(-1, 1))

            Prediction = ('Test MSE: ', mean_squared_error(y_test, y_pred))
            label_Prediction = tk.Label(root, text=Prediction, bg='orange')
            canvas1.create_window(350, 250, window=label_Prediction)

    # Creating 'Calculate Score' button
    button1 = tk.Button(root, text='Calculate Score', command=score, bg='orange')
    canvas1.create_window(215, 200, window=button1)

    
    root.mainloop()

def refresh():
    root.destroy()
    ml_tkinter_gui()

if __name__ == '__main__':
    ml_tkinter_gui()
```

```python
from tkinter import *
from sklearn.ensemble import AdaBoostClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.multiclass import OneVsRestClassifier
import tkinter as tk
from tkinter import ttk
import numpy as np

# Sample data
new_data = {'Face_stress': np.random.rand(100), 'Satisfacation_of_salary': np.random.randint(0, 2, 100)}
X = np.array(new_data['Face_stress']).reshape(-1, 1)
y = np.array(new_data['Satisfacation_of_salary'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)

# Accuracy variables
accuracy_dict = {'SGDClassifier': "0.4285714285714", 'CalibratedClassifierCV':"0.7142857142", 'AdaBoostClassifier': None,
                 'XGBClassifier': None, 'OneVsRestClassifier': None}

# Classifier objects
Alg1 = XGBClassifier()  # Default classifier
Alg2 = AdaBoostClassifier()  # Default classifier
Alg3 = OneVsRestClassifier(SGDClassifier())  # Default classifier
Alg4 = CalibratedClassifierCV()  # Default classifier
Alg6 = OneVsRestClassifier(SGDClassifier())  # Default classifier

def calculate_accuracy(algorithm, model, X_test_data, y_test_data):
    accuracy = model.score(X_test_data.reshape(-1, 1), y_test_data)
    accuracy_dict[algorithm] = accuracy
    print(f"{algorithm} Accuracy: {accuracy}")

def ml_tkinter_gui():
    global root
    root = Tk()

    # Creating canvas to show the result
    canvas1 = tk.Canvas(root, width=500, height=450)
    canvas1.pack()

    # Function to close the window
    def close_window():
        root.destroy()

    # Add button
    button = tk.Button(text="Reference & Restart", command=refresh, bg='green')
    button.place(x=50, y=350)

    # Add button
    button = tk.Button(text="Click and Quit", command=close_window, bg='red')
    button.place(x=310, y=350)

    # Algorithm selection
    Algorithm_list = ['', 'CalibratedClassifierCV', 'AdaBoostClassifier', 'XGBClassifier', 'OneVsRestClassifier']
    Algorithm = StringVar()
    Algorithm.set(Algorithm_list[2])
    Algorithm_combobox = ttk.Combobox(root, width=15, height=10, textvariable=Algorithm)
    Algorithm_combobox['values'] = Algorithm_list
    Algorithm_combobox.place(x=150, y=50)

    # Accuracy Matrices
    def score():
        global s1, Alg3, Alg4, Alg2, Alg1, Alg6
        if Algorithm.get() == 'SGDClassifier':
            Alg3 = OneVsRestClassifier(SGDClassifier())
            Alg3.fit(X_train.reshape(-1, 1), y_train)
            s1 = Alg3.score(X_test.reshape(-1, 1), y_test)
            calculate_accuracy('SGDClassifier', Alg3, X_test, y_test)
        elif Algorithm.get() == 'CalibratedClassifierCV':
            Alg4 = CalibratedClassifierCV()
            Alg4.fit(X_train.reshape(-1, 1), y_train)
            s1 = Alg4.score(X_test.reshape(-1, 1), y_test)
            calculate_accuracy('CalibratedClassifierCV', Alg4, X_test, y_test)
        elif Algorithm.get() == 'AdaBoostClassifier':
            Alg2 = AdaBoostClassifier()
            Alg2.fit(X_train.reshape(-1, 1), y_train)
            s1 = Alg2.score(X_test.reshape(-1, 1), y_test)
            calculate_accuracy('AdaBoostClassifier', Alg2, X_test, y_test)
        elif Algorithm.get() == 'XGBClassifier':
            Alg1 = XGBClassifier()
            Alg1.fit(X_train.reshape(-1, 1), y_train)
            s1 = Alg1.score(X_test.reshape(-1, 1), y_test)
            calculate_accuracy('XGBClassifier', Alg1, X_test, y_test)
        elif Algorithm.get() == 'OneVsRestClassifier':
            Alg6 = OneVsRestClassifier(SGDClassifier())
            Alg6.fit(X_train.reshape(-1, 1), y_train)
            s1 = Alg6.score(X_test.reshape(-1, 1), y_test)
            calculate_accuracy('OneVsRestClassifier', Alg6, X_test, y_test)

        Prediction_result = ('Score: ', s1)
        label_Prediction = tk.Label(root, text=Prediction_result, bg='orange')
        canvas1.create_window(100, 250, window=label_Prediction)

    # Test the data model
    def test_score():
        if 's1' in globals() and s1 is not None:
            if Algorithm.get() == 'OneVsRestClassifier':
                y_pred = Alg6.predict(X_test.reshape(-1, 1))
            elif Algorithm.get() == 'SGDClassifier':
                y_pred = Alg3.predict(X_test.reshape(-1, 1))
            elif Algorithm.get() == 'CalibratedClassifierCV':
                y_pred = Alg4.predict(X_test.reshape(-1, 1))
            elif Algorithm.get() == 'AdaBoostClassifier':
                y_pred = Alg2.predict(X_test.reshape(-1, 1))
            elif Algorithm.get() == 'XGBClassifier':
                y_pred = Alg1.predict(X_test.reshape(-1, 1))

            Prediction = ('Test Scores: ', y_pred)
            label_Prediction = tk.Label(root, text=Prediction, bg='orange')
            canvas1.create_window(350, 250, window=label_Prediction)

    # Creating 'Calculate Score' button
    button1 = tk.Button(root, text='Calculate Score', command=score, bg='orange')
    canvas1.create_window(100, 200, window=button1)

    # Creating 'Calculate Score' button
    button2 = tk.Button(root, text='Calculate test score', command=test_score, bg='orange')
    canvas1.create_window(350, 200, window=button2)

    root.mainloop()

def refresh():
    root.destroy()
    ml_tkinter_gui()

if __name__ == '__main__':
    ml_tkinter_gui()
```
